```python
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
sn.set(style='whitegrid')

plt.figure(figsize=(15, 10))

for i, column in enumerate(df.columns[4:8], 1):
    plt.subplot(2, 3, i)
    sn.histplot(df[column], bins=30, kde=True)
    plt.title(f"Distribution of {column}")
    plt.xlabel(column)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 10))

for i, column in enumerate(df.columns[1:5], 1):
    plt.subplot(2, 3, i)
    sn.boxplot(y=df[column])
    plt.title(f"Distribution of {column}")
plt.tight_layout()
plt.show()
```
